##CenterNet Keypoint Triplets

Bài báo CenterNet là phần phát triển dựa trên CornerNet. CornerNet sử dụng một cặp điểm keypoints để khắc phục nhược điểm của việc sử dụng các phương thức dựa trên khung neo (anchor-based). Tuy nhiên, hiệu suất của CornerNet có khiếm khuyết khi cần nhận dạng ranh giới của các đối tượng vì nó phân tích thông tin toàn cục của đối tượng còn kém. Các tác giả của bài viết CenterNet này đã phân tích hiệu suất của CornerNet. Họ nhận ra rằng tỷ lệ phát hiện sai số của CornerNet trên bộ xác thực MS-COCO là cao (đặc biệt là trên các đối tượng nhỏ) do tỷ lệ của các hộp giới hạn không chính xác.

image.png

CenterNet cố gắng khắc phục các hạn chế gặp phải trong CornerNet. Thay vì dùng thông tin ở 2 góc, nó sử dụng bộ ba để cục bộ hoá lại đối tượng. Thuật toán phát biểu rằng nếu một khung chứa dự đoán (prediction bounding box) có chỉ số IoU so với khung chứa chuẩn (ground-truth bounding box) là cao, thì xác suất điểm chính trung tâm ở khu vực trung tâm của nó được dự đoán là cùng loại cũng cao và ngược lại. Trong quá trình suy luận, đưa ra các điểm góc đề xuất, mô hình sẽ xác minh xem góc đề xuất đí có thực sự là một đối tượng hay không bằng cách kiểm tra xem có một điểm chính trung tâm của cùng một lớp nằm trong khu vực trung tâm của nó hay không.

##Phương pháp Center Pooling

image.png

Một phương pháp Pooling mới được đề xuất để nắm bắt các mẫu hình ảnh phong phú hơn và dễ nhận biết hơn.  Phương pháp này là bắt buộc vì điểm trung tâm của đối tượng không nhất thiết phải truyền tải các mẫu hình ảnh rất dễ nhận biết. Hình trên cho thấy cách pooling thực hiện như thế nào. Đưa ra feature map từ lớp backbone, các tác giả xác định xem một pixel trong feature map có phải là điểm chính giữa hay không. Bản thân pixel trong feature map không chứa đủ thông tin trọng tâm của đối tượng. Do đó, giá trị tối đa của cả hai chiều ngang và dọc được tìm thấy và thêm vào với nhau. Bằng cách này, các tác giả tuyên bố rằng việc phát hiện tốt hơn các điểm chính trung tâm được thực hiện.
